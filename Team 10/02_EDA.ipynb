## Load Data

```python
%pip install seaborn

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Path to  CSV file
file_path = "output_indexed.csv"

# Read the CSV file into a pandas DataFrame
df = pd.read_csv(file_path)

# Display the DataFrame
df.head()
```

```python
df.dtypes
```

```python
# Exclude non-numeric columns and the three columns you want to exclude
columns_to_exclude = ['Description of Requirement', 'Total Base and All Options Value', 'Date Signed']
columns_to_include = [col for col in df.columns if col not in columns_to_exclude]
numeric_columns = df[columns_to_include].select_dtypes(include=['number'])
```

```python
# Create a heatmap
plt.figure(figsize=(10, 8))
heatmap = sns.heatmap(numeric_columns.corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap of Numeric Variables')
plt.show()
```
